### **Problem 1**

We are looking at data from a gamma-ray satellite orbiting in low Earth orbit. It takes a reading of the number of particles detected every 100 milliseconds in an approximately 90-minute orbit. While it is looking for gamma-ray bursts, virtually all of the particles detected are background cosmic rays. 

The data has four columns with more than 25 million rows. Column 1: time (gps seconds). Column 2: solar phase (deg) showing position of sun relative to orbit. Column 3: Earth longitude (deg) showing position of spacecraft relative to ground. Column 4: particle counts.

```python
hf = h5py.File('gammaray_lab4.h5', 'r')
data = np.array(hf.get('data'))
hf.close()
```

```python
time = data[0,:]
sphase = data[1,:]
ephase = data[2,:]
pcount = data[3,:]

plt.hist(ephase, weights = pcount, bins = 1000)
plt.yscale('log')
plt.show()

plt.hist(sphase, weights = pcount, bins = 1000)
plt.yscale('log')
plt.show()

plt.hist(time, weights = ephase, bins = 1000)
plt.yscale('log')
plt.show()

plt.hist(time, weights = sphase, bins = 1000)
plt.yscale('log')
plt.show()

plt.hist(time, weights = pcount, bins = 1000)
plt.yscale('log')
plt.show()
```
